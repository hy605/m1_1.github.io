```python
import requests
# Make a get request for the page
url = "https://www.pewresearch.org/fact-tank/2020/06/11/unemployment-rose-higher-in-three-months-of-covid-19-than-it-did-in-two-years-of-the-great-recession/"
source = requests.get(url)
```

```python
source.content
```

```python
import bs4 as bs 
# Parse the content of the request with BeautifulSoup
page_html = bs.BeautifulSoup(source.content, 'html.parser')
```

```python
# declaring three lists to store data 
titles = []
urls = []
dates = []

# news_containers is a list of "tables"
# For each table
for table in news_containers:

#     find all the hyperlinks by tag "a"
    for link in table.find_all('a'): 
#         print("link:")
#         print(link)
#         print("title:")
#         print(link.text.strip())
        titles.append(link.text.strip()) 
#         print("url:")
#         print(link.get('href'))
        urls.append(link.get('href')) 

#     find all the dates
    for row in table.find_all('tr'):
        date = row.select('div:nth-of-type(2)')[0].text
        print("date")
        print(date)
        dates.append(date)
```

```python
print(len(titles))
print(len(urls))
print(len(dates))
```

```python
# create a dictionary
data2008 = {
    'TITLES': titles,
    'urls': urls,
    'dates': dates
}
```

```python
import pandas as pd 
# create a Pandas frame from a dict 
# guideline: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.from_dict.html
pd_CityUnews2008 = pd.DataFrame.from_dict(data2008)
```
